### Basic Graph

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.linspace(0, 2 * np.pi, 50)  # 0 to 2pi with 50 points
y1 = np.sin(x)
y2 = np.cos(x)

# Resize the graph
plt.figure(figsize=(8, 5), dpi=100)

# Plot sine wave with circle markers, solid blue line
plt.plot(x, y1, 'bo-', label='sin(x)', linewidth=2, markersize=6)

# Plot cosine wave with square markers, dashed red line
plt.plot(x, y2, 'rs--', label='cos(x)', linewidth=2, markersize=6)

# Title and labels
plt.title('Sine and Cosine Waves', fontdict={'fontsize': 18, 'fontweight': 'bold'})
plt.xlabel('X Axis (radians)')
plt.ylabel('Y Axis (value)')

# Tick marks for x and y
plt.xticks(np.arange(0, 2 * np.pi + 0.1, np.pi / 2),
           ['0', '/2', '', '3/2', '2'])
plt.yticks(np.arange(-1, 1.5, 0.5))

# Legend
plt.legend()

# Save the figure at high resolution
plt.savefig('sine_cosine_plot.png', dpi=300)

# Show the plot
plt.show()
```

### Bar Chart

```python


labels = ['Apple', 'Banana', 'Cherry']
values = [10, 7, 5]

plt.figure(figsize=(5, 3), dpi=100)

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.title("Fruit Sales")
plt.ylabel("Quantity Sold")
plt.savefig('fruit_sales_barchart.png', dpi=300)

plt.show()
```

### Line Graph

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('co2_emission.csv')

# Clean column names: remove extra spaces
df.columns = df.columns.str.strip()

# Filter relevant countries using the correct column name: 'Entity'
countries_to_plot = ['United States', 'China', 'India', 'Russia', 'Japan']
df_filtered = df[df['Entity'].isin(countries_to_plot)]

# Pivot data for plotting
df_pivot = df_filtered.pivot(index='Year', columns='Entity', values='Annual CO emissions (tonnes )')

# Plotting
plt.figure(figsize=(10,6))
plt.title('CO Emissions Over Time (in Tonnes)', fontdict={'fontweight':'bold', 'fontsize': 18})

# Custom plotting style for each country
colors = ['b', 'r', 'g', 'y', 'm']
for i, country in enumerate(countries_to_plot):
    plt.plot(df_pivot.index, df_pivot[country], f'{colors[i]}.-', label=country)

# Axis labels and ticks
plt.xlabel('Year')
plt.ylabel('Annual CO Emissions (tonnes)')
plt.xticks(df_pivot.index[::5].tolist() + [2021])
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save the figure (optional)
plt.savefig('co2_emissions_over_time.png', dpi=300)

# Show plot
plt.show()
```

### Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('train.csv')  # Titanic training data

# Inspect columns
print(df.columns)

# Clean: drop missing ages
ages = df['Age'].dropna()

# Plot histogram
plt.figure(figsize=(10,6))
plt.hist(ages, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Passenger Ages on Titanic', fontdict={'fontsize':16, 'fontweight':'bold'})
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

### Pie Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('fruit_sales.csv')

# Plot pie chart
plt.figure(figsize=(8,8))
plt.pie(df['Sales'], labels=df['Fruit'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('Fruit Sales Distribution')
plt.axis('equal')  # Equal aspect ratio to ensure it's a circle
plt.show()
```

### Pie Chart #2

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Titanic data
df = pd.read_csv('train.csv')  # Make sure to download from Kaggle

# Count number of males and females
gender_counts = df['Sex'].value_counts()

# Plot pie chart
plt.figure(figsize=(8,8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], startangle=90)
plt.title('Gender Distribution on the Titanic')
plt.axis('equal')
plt.show()
```

### Box and Whiskers Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="matplotlib")


# Load Titanic training data from Kaggle file
titanic = pd.read_csv('train.csv')

# Extract ages by passenger class, dropping missing ages
class1 = titanic.loc[titanic.Pclass == 1, 'Age'].dropna()
class2 = titanic.loc[titanic.Pclass == 2, 'Age'].dropna()
class3 = titanic.loc[titanic.Pclass == 3, 'Age'].dropna()

plt.figure(figsize=(6,8), dpi=100)
plt.style.use('default')

bp = plt.boxplot(
    [class1, class2, class3],
    tick_labels=['1st Class', '2nd Class', '3rd Class'],
    patch_artist=True,
    medianprops={'linewidth': 2, 'color': 'darkred'}
)

plt.title('Age Distribution by Passenger Class on Titanic')
plt.ylabel('Age (years)')

for box in bp['boxes']:
    box.set(color='#1f77b4', linewidth=2)
    box.set(facecolor='#d0e1f9')

plt.show()
```
